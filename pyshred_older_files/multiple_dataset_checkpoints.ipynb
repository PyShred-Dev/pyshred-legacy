```python
# Generate sample data
# row = 256 -> 200 (slicing) -> 50 (resizing w/ cubic interpolation)
# columns = 257 -> 240 (slicing) -> 60 (resizing w/ cubic interpolation)
# timesteps = 2000

plasma_data_Jez = sio.loadmat('data14fields/Jez')['Data']
plasma_data_Jez = plasma_data_Jez.reshape(257,256,2000)
plasma_data_Jez = np.transpose(plasma_data_Jez, axes=(1, 0, 2))
plasma_data_Jez = plasma_data_Jez[28:228,8:248,:]
scale_factors = (50 / plasma_data_Jez.shape[0], 60 / plasma_data_Jez.shape[1], 1)
plasma_data_Jez = zoom(plasma_data_Jez, zoom=scale_factors)

plasma_data_Jey = sio.loadmat('data14fields/Jey')['Data']
plasma_data_Jey = plasma_data_Jey.reshape(257,256,2000)
plasma_data_Jey = np.transpose(plasma_data_Jey, axes=(1, 0, 2))
plasma_data_Jey = plasma_data_Jey[28:228,8:248,:]
plasma_data_Jey = zoom(plasma_data_Jey, zoom=scale_factors)

plasma_data_Jex = sio.loadmat('data14fields/Jex')['Data']
plasma_data_Jex = plasma_data_Jex.reshape(257,256,2000)
plasma_data_Jex = np.transpose(plasma_data_Jex, axes=(1, 0, 2))
plasma_data_Jex = plasma_data_Jex[28:228,8:248,:]
plasma_data_Jex = zoom(plasma_data_Jex, zoom=scale_factors, order = 3)


np.savez_compressed('plasma_data', Jex = plasma_data_Jex, Jey = plasma_data_Jey, Jez = plasma_data_Jez)
```
